```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_df = pd.read_csv('company_sales_data.csv')
pd_df.head(12)
```

```python
# 1.Прочитайте все данные о продажах продукта и покажите их с помощью многострочного графика  
month_number_list = pd_df['month_number']
facecream_list = pd_df['facecream']
facewash_list = pd_df['facewash']
toothpaste_list = pd_df['toothpaste']
bathingsoap_list = pd_df['bathingsoap']
shampoo_list = pd_df['shampoo']
moisturizer_list = pd_df['moisturizer']
```

```python
plt.figure(figsize = (12, 8))
plt.plot(month_number_list, facecream_list, 
         label = 'Крем для лица', 
         color = 'b', 
         linestyle = '--',
         linewidth = 1,
         marker = 'o', 
         markerfacecolor = 'b' )

plt.plot(month_number_list, facewash_list, 
         label = 'Средство для умывания', 
         color = 'g', 
         linestyle = '--',
         linewidth = 1,
         marker = 'o', 
         markerfacecolor = 'g' )

plt.plot(month_number_list, toothpaste_list, 
         label = 'Зубная паста', 
         color = 'r', 
         linestyle = '--',
         linewidth = 1,
         marker = 'o', 
         markerfacecolor = 'r' )

plt.plot(month_number_list, bathingsoap_list, 
         label = 'Мыло для купания', 
         color = 'c', 
         linestyle = '--',
         linewidth = 1,
         marker = 'o', 
         markerfacecolor = 'c' )

plt.plot(month_number_list, shampoo_list, 
         label = 'Шампунь', 
         color = 'm', 
         linestyle = '--',
         linewidth = 1,
         marker = 'o', 
         markerfacecolor = 'm' )

plt.plot(month_number_list, moisturizer_list, 
         label = 'Средство для увлажнения', 
         color = 'y', 
         linestyle = '--',
         linewidth = 1,
         marker = 'o', 
         markerfacecolor = 'y' )

plt.xticks(month_number_list)
plt.title('Информация о продажах за последний год')
plt.ylabel('Количество проданных единиц')
plt.xlabel('Номер месяца')
plt.grid()
plt.legend()
plt.show()
```

```python
# 2.Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью гистограммы.

plt.figure(figsize = (12, 8))
plt.hist(bathingsoap_list, bins = 12,
#          density = 1,
         label = 'Мыло для купания', 
         color = 'c')
plt.title('Информация о продажах мыла для купания за последний год')
plt.legend()
plt.show()
```

```python
# 3.Рассчитайте общие данные о продажах за прошлый год для каждого продукта и покажите их с помощью круговой диаграмsuмы
product_names = ['facecream', 'facewash', 'toothpaste', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
product_data = [sum(month_number_list), sum(facecream_list), sum(facewash_list), 
                sum(toothpaste_list), sum(bathingsoap_list), sum(shampoo_list), sum(moisturizer_list)]

plt.figure(figsize = (12, 8))
plt.pie(product_data, labels = product_names)
plt.title('Информация о продажах всех товаров за последний год')
plt.show()
```
